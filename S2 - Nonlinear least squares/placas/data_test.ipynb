```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
im = Image.open("output.png").convert('L')
im
```

```python
# invert image
im = Image.eval(im, lambda x: 255-x)
im
```

```python
# crop left and right corners (remove a little bit of both)
im = im.crop((20, 20, im.width-35, im.height - 5))
im
```

```python
# Create 7 new images by "rolling" a 24x40 window over the image, left to right with a 3-pixel overlap
ims = []
for i in range(7):
    start_x = i * (24 - 3)  # Overlap 3 pixels
    ims.append(im.crop((start_x, 0, start_x + 24, 40)))
```

```python
# plot images
fig, axs = plt.subplots(1, 7, figsize=(20, 5))
for i, ax in enumerate(axs):
    ax.imshow(ims[i], cmap='gray')
    ax.axis('off')
plt.show()
```
