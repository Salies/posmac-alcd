```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import truncnorm
```

```python
df = pd.read_csv('data/imdb_top_1000.csv')
# remove repeated Series_Title
df = df.drop_duplicates(subset='Series_Title', keep='first')
print(len(df))
df.head()
```

```python
np.random.seed(7)
# get 100 random movies
df = df.sample(100)
movie_sample = df['Series_Title'].values
movie_sample
```

```python
np.random.seed(6661)

mean = 5  # center of the range
std_dev = 2  # controls the spread
lower, upper = 0, 10  # range boundaries

user_ratings = []

for i in range(100):
    numbers = truncnorm((lower - mean) / std_dev, (upper - mean) / std_dev, loc=mean, scale=std_dev).rvs(100)
    # round to 0.5
    numbers = np.round(numbers * 2) / 2
    user_ratings.append(numbers)
    sns.histplot(numbers, kde=True)
numbers
```

```python
users = np.array([f'User{x}' for x in range(1, 101)])
users
```

```python
# create df where columns are users and rows are movies
df_out = pd.DataFrame(user_ratings, columns=users)
df_out['Movie'] = movie_sample
df_out = df_out.set_index('Movie')
df_out.head()
```

```python
# save as movie_user_data.csv
df_out.to_csv('data/movie_user_data.csv')
```
