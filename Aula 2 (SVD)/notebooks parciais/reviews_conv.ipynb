```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('data/Video_Games.json', lines=True)
df.head()
```

```python
df['overall'].value_counts()
```

```python
# restricit to the 500 most active users
user_counts = df['reviewerID'].value_counts()
top_users = user_counts.head(500)
top_users = top_users.index
df = df[df['reviewerID'].isin(top_users)]
```

```python
pivot_table = df.pivot_table(index='asin', columns='reviewerID', values='overall')
pivot_table.head()
```

Há muitos usuários que não fizeram review dos mesmos jogos. Vamos considerar como 0.

```python
pivot_table.fillna(0, inplace=True)
pivot_table.head()
```

```python
# count zeroes
((pivot_table == 0).sum()).sum()
```
